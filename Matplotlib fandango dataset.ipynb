```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
reviews = pd.read_csv("fandango_scores.csv")
```

```python
norm_reviews = reviews[["FILM", "RT_user_norm", "Metacritic_user_nom", "IMDB_norm", "Fandango_Ratingvalue", "Fandango_Stars"]]
norm_reviews.head()
```

Average grade for Avengers : Age of Ultron


```python
fig, ax = plt.subplots()

left = np.arange(5) + 1
col_names = ["RT_user_norm", "Metacritic_user_nom", "IMDB_norm", "Fandango_Ratingvalue", "Fandango_Stars"]

height = norm_reviews[col_names].iloc[0].values
ax.bar(left, height, 0.5)

ax.set_xticks(range(1,6))
ax.set_xticklabels(col_names, rotation=90)

ax.set_xlabel("Source Notation")
ax.set_ylabel("Average Grade")
ax.set_title("Average grade for the movie Avengers: Age of Ultron (2015)")

plt.show()
```

Ant-Man's average grades

```python
fig, ax = plt.subplots()

y_axe = np.arange(5) + 1
x_axe = norm_reviews[col_names].iloc[2].values

ax.barh(y_axe, x_axe, 0.5)

ax.set_yticks(range(1,6))
ax.set_yticklabels(col_names)

ax.set_title("Average grade for the movie Ant-Man (2015)")
ax.set_ylabel("Source Notation",)
ax.set_xlabel("Average grade")


plt.show()
```

```python
fig, ax = plt.subplots()

ax.scatter(norm_reviews["Fandango_Ratingvalue"], norm_reviews["RT_user_norm"])

ax.set_xlabel("Fandango")
ax.set_ylabel("Rotten Tomatos")

plt.show()
```

```python
fig = plt.figure(figsize=(15,4))

ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

ax1.scatter(norm_reviews["Fandango_Ratingvalue"], norm_reviews["RT_user_norm"])
ax2.scatter(norm_reviews["RT_user_norm"], norm_reviews["Fandango_Ratingvalue"])

ax1.set_xlabel("Fandango")
ax2.set_xlabel("Rotten tomatos")

ax1.set_ylabel("Rottent tomatos")
ax2.set_ylabel("Fandango")

plt.show()
```

```python
fig = plt.figure(figsize=(10,5))

ax1 = fig.add_subplot(2,2,1)
ax1.scatter(norm_reviews["Fandango_Ratingvalue"], norm_reviews["RT_user_norm"])
ax1.set_xlabel("Fandango")
ax1.set_ylabel("Rotten tomatos")
ax1.set_xlim(0,5)
ax1.set_ylim(0,5)

ax2 = fig.add_subplot(2,2,2)
ax2.scatter(norm_reviews["Fandango_Ratingvalue"], norm_reviews["Metacritic_user_nom"])
ax2.set_xlabel("Fandango")
ax2.set_ylabel("Metacritic")
ax2.set_xlim(0,5)
ax2.set_ylim(0,5)

ax3 = fig.add_subplot(2,2,3)
ax3.scatter(norm_reviews["Fandango_Ratingvalue"], norm_reviews["IMDB_norm"])
ax3.set_xlabel("Fandango")
ax3.set_ylabel("IMDB")
ax3.set_xlim(0,5)
ax3.set_ylim(0,5)

plt.show()
```

```python
fig = plt.figure(figsize=(12,9))

ax1 = fig.add_subplot(2,2,1)
ax1.hist(norm_reviews["Fandango_Ratingvalue"], bins=20, range=(0,5))
ax1.set_ylim(0,50)
ax1.set_title("Fandango")

ax2 = fig.add_subplot(2,2,2)
ax2.hist(norm_reviews["RT_user_norm"], bins=20, range=(0,5))
ax2.set_ylim(0,50)
ax2.set_title("Rotten tomatos")

ax3 = fig.add_subplot(2,2,3)
ax3.hist(norm_reviews["Metacritic_user_nom"], bins=20, range=(0,5))
ax3.set_ylim(0,50)
ax3.set_title("Metacritic")

ax4 = fig.add_subplot(2,2,4)
ax4.hist(norm_reviews["IMDB_norm"], bins=20, range=(0,5))
ax4.set_ylim(0,50)
ax4.set_title("IMDB")


plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6,3))

col_names = ["Fandango_Ratingvalue", "Metacritic_user_nom", "RT_user_norm", "IMDB_norm"]

ax.boxplot(norm_reviews[col_names].values)
ax.set_ylim(0,5)
ax.set_xticks(range(1,5))
ax.set_xticklabels(col_names, rotation=90)


plt.show()
```
